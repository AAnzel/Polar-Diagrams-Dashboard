```python
path_wine_root = os.path.join('..', '..', 'data', 'Case_Study_Wine')
path_wine_cluttered_root = os.path.join(
    '..', '..', 'data', 'Case_Study_Cluttered_Wine')
path_wine_original = os.path.join(path_wine_root, 'winequality-red.csv')
path_wine_result = os.path.join(path_wine_root, 'wine_sampled.csv')
path_cluttered_wine_result = os.path.join(
    path_wine_cluttered_root, 'wine_cluttered.csv')
df_wine_original = pd.read_csv(path_wine_original, sep=';')
df_wine_original
```

```python
df_wine_original.dropna(inplace=True)
df_wine_original
```

```python
df_wine_sampled = df_wine_original.groupby('quality').apply(
    lambda x: x.sample(frac=0.0125)).droplevel(0)
df_wine_sampled
```

```python
df_cluttered_wine_sampled = df_wine_original.head(50)
df_cluttered_wine_sampled
```

```python
#df_wine_sampled = df_wine_sampled.drop('quality', axis=1)
df_wine_sampled['Wine'] = ['Wine_' + str(i)
                           for i in range(df_wine_sampled.shape[0])]
df_wine_sampled
```

```python
#df_wine_sampled = df_wine_sampled.drop('quality', axis=1)
df_cluttered_wine_sampled['Wine'] = ['Wine_' + str(i)
                           for i in range(df_cluttered_wine_sampled.shape[0])]
df_cluttered_wine_sampled
```

```python
df_wine_sampled_pivoted = df_wine_sampled.pivot_table(columns='Wine')
df_wine_sampled_pivoted
```

```python
df_cluttered_wine_sampled_pivoted = df_cluttered_wine_sampled.pivot_table(columns='Wine')
df_cluttered_wine_sampled_pivoted
```

```python
series_median = df_wine_sampled[df_wine_sampled.drop('Wine', axis=1).columns.to_list()].median()
series_median
```

```python
series_cluttered_median = df_cluttered_wine_sampled[df_cluttered_wine_sampled.drop('Wine', axis=1).columns.to_list()].median()
series_cluttered_median
```

```python
df_wine_sampled_pivoted['Median'] = series_median
df_wine_sampled_pivoted
```

```python
df_cluttered_wine_sampled_pivoted['Median'] = series_cluttered_median
df_cluttered_wine_sampled_pivoted
```

```python
df_wine_sampled_pivoted.to_csv(path_wine_result, index=False)
df_cluttered_wine_sampled_pivoted.to_csv(path_cluttered_wine_result, index=False)
```
